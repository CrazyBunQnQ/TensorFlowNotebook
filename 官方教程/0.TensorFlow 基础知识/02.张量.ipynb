`tf.Tensor` åŸºç±»è¦æ±‚å¼ é‡æ˜¯â€œçŸ©å½¢â€â€”â€”ä¹Ÿå°±æ˜¯è¯´ï¼Œæ¯ä¸ªè½´ä¸Šçš„æ¯ä¸€ä¸ªå…ƒç´ å¤§å°ç›¸åŒã€‚ä½†æ˜¯ï¼Œå¼ é‡æœ‰å¯ä»¥å¤„ç†ä¸åŒå½¢çŠ¶çš„ç‰¹æ®Šç±»å‹ã€‚

- ä¸è§„åˆ™å¼ é‡ï¼ˆå‚é˜…ä¸‹æ–‡ä¸­çš„ [RaggedTensor](#ragged_tensors)ï¼‰
- ç¨€ç–å¼ é‡ï¼ˆå‚é˜…ä¸‹æ–‡ä¸­çš„ [SparseTensor](#sparse_tensors)ï¼‰

æ‚¨å¯ä»¥å¯¹å¼ é‡æ‰§è¡ŒåŸºæœ¬æ•°å­¦è¿ç®—ï¼ŒåŒ…æ‹¬åŠ æ³•ã€é€å…ƒç´ ä¹˜æ³•å’Œ[çŸ©é˜µä¹˜æ³•](https://baike.baidu.com/item/%E7%9F%A9%E9%98%B5%E4%B9%98%E6%B3%95/5446029)ã€‚

````python
a = tf.constant([[1, 2],
                 [3, 4]])
b = tf.constant([[1, 1],
                 [1, 1]]) # ä¹Ÿå¯ä»¥è¯´ `tf.ones([2,2])`

print(tf.add(a, b), "\n")
print(tf.multiply(a, b), "\n")
# çŸ©é˜µä¹˜æ³•(matrix multiplication)
print(tf.matmul(a, b), "\n")
````

````python
print(a + b, "\n") # é€å…ƒç´ åŠ æ³•
print(a * b, "\n") # é€å…ƒç´ ä¹˜æ³•
print(a @ b, "\n") # çŸ©é˜µä¹˜æ³•(matrix multiplication)
````

- [`tf.add(x, y)`](https://www.tensorflow.org/api_docs/python/tf/math/add): x + y, è‹¥ä¸€ä¸ªæ˜¯å¼ é‡ï¼Œä¸€ä¸ªæ˜¯éå¼ é‡ï¼Œåˆ™ç»“æœä¸ºå¼ é‡

    ```python

        x, y, name=None
    )
    ```
- [`tf.matmul(a, b)`](https://www.tensorflow.org/api_docs/python/tf/linalg/matmul): å°†çŸ©é˜µ a ä¸çŸ©é˜µ b ç›¸ä¹˜ï¼Œå¾—åˆ° a * b

å„ç§è¿ç®— (op) éƒ½å¯ä»¥ä½¿ç”¨å¼ é‡ã€‚

````python
c = tf.constant([[4.0, 5.0], [10.0, 1.0]])
print(c)

# æ‰¾åˆ°æœ€å¤§å€¼
print(tf.reduce_max(c))
# æ‰¾åˆ°æœ€å¤§å€¼çš„ç´¢å¼•
print(tf.argmax(c))
# è®¡ç®—softmax(Compute the softmax)
print(tf.nn.softmax(c))
````
